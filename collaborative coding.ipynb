```python
import pandas as pd
import numpy as np
from sklearn.decomposition import NMF

# Read data from CSV file
file_path = r"C:\Users\User\Desktop\response.csv"
data = pd.read_csv(r"C:\Users\User\Desktop\response.csv")

# Preprocess the data if necessary
# For example, handle missing values or clean the data

# Create User-Item Matrix for Johor
user_item_matrix_johor = pd.pivot_table(data, values=['traffic\n', 'air_pollution\n', 'sound_pollution', 'crime_rate', 'overall_rating'], index='Username')
# Create User-Item Matrix for Melaka
user_item_matrix_melaka = pd.pivot_table(data, values=['traffic\n.1', 'air_pollution\n.1', 'sound_pollution.1', 'crime_rate.1', 'overall_rating.1'], index='Username')

# Fill missing values with 0 for Melaka
user_item_matrix_melaka.fillna(0, inplace=True)

# Convert Melaka user-item matrix to numpy array
X_melaka = user_item_matrix_melaka.values

# Initialize and fit NMF model for Melaka
num_components = 20  # Number of latent factors
model_melaka = NMF(n_components=num_components, init='random', random_state=20)
W_melaka = model_melaka.fit_transform(X_melaka)  # Factorization of X_melaka into W_melaka and H_melaka

# Reconstruction of X from the factorized matrices
X_reconstructed_melaka = np.dot(W_melaka, model_melaka.components_)

# Round the reconstructed matrix to 1 decimal place
X_reconstructed_rounded_melaka = np.round(X_reconstructed_melaka, 1)

# Print the reconstructed matrix with numbers rounded to 1 decimal place
print("\nReconstructed Matrix for Melaka:")
print(pd.DataFrame(X_reconstructed_rounded_melaka, index=user_item_matrix_melaka.index, columns=user_item_matrix_melaka.columns))

# Calculate RMSE
rmse_melaka = np.sqrt(np.mean((X_melaka - X_reconstructed_melaka)**2))
print("\nRoot Mean Square Error (RMSE) for Melaka:", rmse_melaka)

# Repeat the same process for Negeri Sembilan
user_item_matrix_negeri_sembilan = pd.pivot_table(data, values=['traffic\n.2', 'air_pollution\n.2', 'sound_pollution.2', 'crime_rate.2', 'overall_rating.2'], index='Username')
user_item_matrix_negeri_sembilan.fillna(0, inplace=True)
X_negeri_sembilan = user_item_matrix_negeri_sembilan.values
model_negeri_sembilan = NMF(n_components=num_components, init='random', random_state=20)
W_negeri_sembilan = model_negeri_sembilan.fit_transform(X_negeri_sembilan)
X_reconstructed_negeri_sembilan = np.dot(W_negeri_sembilan, model_negeri_sembilan.components_)
X_reconstructed_rounded_negeri_sembilan = np.round(X_reconstructed_negeri_sembilan, 1)
print("\nReconstructed Matrix for Negeri Sembilan:")
print(pd.DataFrame(X_reconstructed_rounded_negeri_sembilan, index=user_item_matrix_negeri_sembilan.index, columns=user_item_matrix_negeri_sembilan.columns))
rmse_negeri_sembilan = np.sqrt(np.mean((X_negeri_sembilan - X_reconstructed_negeri_sembilan)**2))
print("\nRoot Mean Square Error (RMSE) for Negeri Sembilan:", rmse_negeri_sembilan)


# Fill missing values with 0 for Johor
user_item_matrix_johor.fillna(0, inplace=True)

# Convert Johor user-item matrix to numpy array
X_johor = user_item_matrix_johor.values

# Initialize and fit NMF model for Johor
num_components = 20  # Number of latent factors
model_johor = NMF(n_components=num_components, init='random', random_state=20)
W_johor = model_johor.fit_transform(X_johor)  # Factorization of X_johor into W_johor and H_johor

# Reconstruction of X from the factorized matrices
X_reconstructed_johor = np.dot(W_johor, model_johor.components_)

# Round the reconstructed matrix to 1 decimal place
X_reconstructed_rounded_johor = np.round(X_reconstructed_johor, 1)

# Print the reconstructed matrix with numbers rounded to 1 decimal place
print("Reconstructed Matrix for Johor:")
print(pd.DataFrame(X_reconstructed_rounded_johor, index=user_item_matrix_johor.index, columns=user_item_matrix_johor.columns))

# Calculate RMSE for Johor
rmse_johor = np.sqrt(np.mean((X_johor - X_reconstructed_johor)**2))
print("Root Mean Square Error (RMSE) for Johor:", rmse_johor)
```

```python
import pandas as pd
import numpy as np
from surprise import Reader, Dataset, KNNBasic
from surprise import accuracy
from surprise.model_selection import train_test_split

# Read data from CSV file
file_path = r"C:\Users\User\Desktop\response.csv"
data = pd.read_csv(file_path)

# Preprocess the data if necessary
# For example, handle missing values or clean the data

# Helper function to prepare the data for the Surprise library
def prepare_data_for_surprise(data, state_columns, user_column='Username'):
    df = data.melt(id_vars=[user_column], value_vars=state_columns,
                   var_name='item', value_name='rating')
    df.dropna(subset=['rating'], inplace=True)
    df['item'] = df['item'].astype(str)
    return df[[user_column, 'item', 'rating']]

# Prepare data for Melaka
state_columns_melaka = ['traffic_melaka\n', 'air_pollution_melaka\n', 'sound_pollution_melaka', 'crime_rate_melaka', 'overall_rating_melaka']
melaka_data = prepare_data_for_surprise(data, state_columns_melaka)

# Define the reader for the Surprise library
reader = Reader(rating_scale=(0, 5))

# Load the data into Surprise dataset
melaka_dataset = Dataset.load_from_df(melaka_data, reader)

# Split the data into training and test sets
trainset, testset = train_test_split(melaka_dataset, test_size=0.2, random_state=20)

# Define the KNN model for item-based collaborative filtering
sim_options = {'name': 'cosine', 'user_based': False}  # Use cosine similarity and item-based CF
knn_model = KNNBasic(sim_options=sim_options)

# Train the model
knn_model.fit(trainset)

# Predict ratings for the test set
predictions = knn_model.test(testset)

# Calculate RMSE
rmse_melaka = accuracy.rmse(predictions)
print("\nRoot Mean Square Error (RMSE) for Melaka:", rmse_melaka)

# Repeat the same process for Negeri Sembilan
state_columns_ns = ['traffic_ns\n', 'air_pollution_ns\n', 'sound_pollution_ns', 'crime_rate_ns', 'overall_rating_ns']
ns_data = prepare_data_for_surprise(data, state_columns_ns)
ns_dataset = Dataset.load_from_df(ns_data, reader)
trainset, testset = train_test_split(ns_dataset, test_size=0.2, random_state=20)
knn_model.fit(trainset)
predictions = knn_model.test(testset)
rmse_ns = accuracy.rmse(predictions)
print("\nRoot Mean Square Error (RMSE) for Negeri Sembilan:", rmse_ns)

# Repeat the same process for Johor
state_columns_johor = ['traffic\n', 'air_pollution_johor\n', 'sound_pollution_johor', 'crime_rate_johor', 'overall_rating_johor']
johor_data = prepare_data_for_surprise(data, state_columns_johor)
johor_dataset = Dataset.load_from_df(johor_data, reader)
trainset, testset = train_test_split(johor_dataset, test_size=0.2, random_state=20)
knn_model.fit(trainset)
predictions = knn_model.test(testset)
rmse_johor = accuracy.rmse(predictions)
print("\nRoot Mean Square Error (RMSE) for Johor:", rmse_johor)
```

```python
import pandas as pd
import numpy as np
from surprise import Reader, Dataset, KNNBasic
from surprise import accuracy
from surprise.model_selection import train_test_split

# Read data from CSV file
file_path = r"C:\Users\User\Desktop\response.csv"
data = pd.read_csv(file_path)

# Helper function to prepare the data for the Surprise library
def prepare_data_for_surprise(data, state_columns, user_column='Username'):
    df = data.melt(id_vars=[user_column], value_vars=state_columns,
                   var_name='item', value_name='rating')
    df.dropna(subset=['rating'], inplace=True)
    df['item'] = df['item'].astype(str)
    return df[[user_column, 'item', 'rating']]

# Prepare data for Melaka
state_columns_melaka = ['traffic_melaka\n', 'air_pollution_melaka\n', 'sound_pollution_melaka', 'crime_rate_melaka', 'overall_rating_melaka']
melaka_data = prepare_data_for_surprise(data, state_columns_melaka)

# Define the reader for the Surprise library
reader = Reader(rating_scale=(0, 5))

# Load the data into Surprise dataset
melaka_dataset = Dataset.load_from_df(melaka_data, reader)

# Split the data into training and test sets
trainset, testset = train_test_split(melaka_dataset, test_size=0.2, random_state=20)

# Define the KNN model for item-based collaborative filtering
sim_options = {'name': 'cosine', 'user_based': False}  # Use cosine similarity and item-based CF
knn_model = KNNBasic(sim_options=sim_options)

# Train the model
knn_model.fit(trainset)

# Predict ratings for the test set
predictions = knn_model.test(testset)

# Calculate RMSE
rmse_melaka = accuracy.rmse(predictions)
print("\nRoot Mean Square Error (RMSE) for Melaka:", rmse_melaka)

# Function to calculate cosine similarity matrix for items
def calculate_cosine_similarity(trainset):
    num_items = trainset.n_items
    num_users = trainset.n_users
    item_user_matrix = np.zeros((num_items, num_users))

    for item_id in range(num_items):
        for user_id, rating in trainset.ir[item_id]:
            item_user_matrix[item_id, user_id] = rating

    # Normalize the item-user matrix
    norm_item_user_matrix = item_user_matrix / np.linalg.norm(item_user_matrix, axis=1)[:, np.newaxis]

    # Calculate cosine similarity
    similarity_matrix = np.dot(norm_item_user_matrix, norm_item_user_matrix.T)
    return similarity_matrix

# Calculate and print the cosine similarity matrix for items in Melaka
cosine_similarity_melaka = calculate_cosine_similarity(trainset)
print("\nCosine Similarity Matrix for Melaka:")
print(pd.DataFrame(cosine_similarity_melaka))

# Repeat the same process for Negeri Sembilan
state_columns_ns = ['traffic_ns\n', 'air_pollution_ns\n', 'sound_pollution_ns', 'crime_rate_ns', 'overall_rating_ns']
ns_data = prepare_data_for_surprise(data, state_columns_ns)
ns_dataset = Dataset.load_from_df(ns_data, reader)
trainset, testset = train_test_split(ns_dataset, test_size=0.2, random_state=20)
knn_model.fit(trainset)
predictions = knn_model.test(testset)
rmse_ns = accuracy.rmse(predictions)
print("\nRoot Mean Square Error (RMSE) for Negeri Sembilan:", rmse_ns)
cosine_similarity_ns = calculate_cosine_similarity(trainset)
print("\nCosine Similarity Matrix for Negeri Sembilan:")
print(pd.DataFrame(cosine_similarity_ns))

# Repeat the same process for Johor
state_columns_johor = ['traffic\n', 'air_pollution_johor\n', 'sound_pollution_johor', 'crime_rate_johor', 'overall_rating_johor']
johor_data = prepare_data_for_surprise(data, state_columns_johor)
johor_dataset = Dataset.load_from_df(johor_data, reader)
trainset, testset = train_test_split(johor_dataset, test_size=0.2, random_state=20)
knn_model.fit(trainset)
predictions = knn_model.test(testset)
rmse_johor = accuracy.rmse(predictions)
print("\nRoot Mean Square Error (RMSE) for Johor:", rmse_johor)
cosine_similarity_johor = calculate_cosine_similarity(trainset)
print("\nCosine Similarity Matrix for Johor:")
print(pd.DataFrame(cosine_similarity_johor))
```
